```python
N_theta, N_phi = 100, 100
theta = np.linspace(0, np.pi, N_theta)
phi = np.linspace(0, 2*np.pi, N_phi, endpoint=False)
dtheta, dphi = theta[1] - theta[0], phi[1] - phi[0]

theta_grid, phi_grid = np.meshgrid(theta, phi, indexing='ij')

frequencies = np.linspace(30e9, 300e9, 10)  # Frequencies from 30 GHz to 300 GHz
aspect_ratios = np.array([0.1, 0.5, 1, 2, 3])  # Aspect ratios to test

A_h, A_v = compute_polarizability(frequencies, aspect_ratios)

elevation = 66 # degrees

delta = np.radians(elevation)  # Convert elevation to radians
```

```python
A_h_exp = A_h[:, :, None, None]
A_v_exp = A_v[:, :, None, None]
theta_exp = theta_grid[None, None, :, :]
phi_exp = phi_grid[None, None, :, :]

M_sc = compute_earth_frame_phase_matrix(A_h_exp, A_v_exp, theta_exp, phi_exp, delta)
M_tensor = compute_scattering_plane_phase_matrix(A_h_exp, A_v_exp, theta_exp, phi_exp, delta)
```

```python
# Let's compare M_sc and M_tensor as we should be getting the same results, we'll do a boolian check to see if they are close enough
print("Are M_sc and M_tensor close enough?", np.allclose(M_sc, M_tensor, atol=1e-6))
```
