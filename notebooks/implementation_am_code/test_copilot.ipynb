```python
import pycraf 
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u
from scipy.integrate import trapezoid
```

```python
altitudes = np.geomspace(1, 25000, 1000) + 4999 #m


altitudes_km = altitudes * u.m       # maintenant c'est une Quantity en m
altitudes_km = altitudes_km.to(u.km) # conversion en km
Temperature = pycraf.atm.profile_standard(altitudes_km)[0].value #en K
Pressure = pycraf.atm.profile_standard(altitudes_km)[1].value #en hPa
rho_water = pycraf.atm.profile_standard(altitudes_km)[2].value #en g/m3
P_water = pycraf.atm.profile_standard(altitudes_km)[3].value #en hPa
```

```python
#Let's estimate the total air mass in the atmosphere by integrating the density profile over altitude. We can use the trapezoidal rule for numerical integration.

# Convert to rho_air in kg/m^3
rho_air = Pressure * 100 / (Temperature * 287.05) # using ideal gas law, where R = 287.05 J/(kg*K)
# Integrate the density profile to get the total mass per unit area (kg/m^2)
total_mass_per_unit_area = trapezoid(rho_air, altitudes)
print(f"Total mass of air per unit area: {total_mass_per_unit_area:.2f} kg/m^2")
```
